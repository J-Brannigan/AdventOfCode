```python
import numpy as np
```

```python
with open('puzzle_input.txt') as f:
    first_line = f.readline()
    print(first_line)
```

```python
range_strings = first_line.split(',')
print(range_strings[1])
```

```python
ranges=[]
for range_string in range_strings:
    ranges.append({"start_num": int(range_string.split('-')[0]), "end_num":int(range_string.split('-')[1])})
print(ranges)
```

```python
def is_valid_id(id):
    id_str = str(id)
    if len(id_str)%2!=0:
        return True
    else:
        first_half = id_str[:int(len(id_str)/2)]
        second_half = id_str[int(len(id_str)/2):]
        if first_half == second_half:
            return False
        else:
            return True
```

```python
total = 0
for range_dict in ranges:
    for id in np.arange(range_dict['start_num'],range_dict['end_num']+1):
        if not is_valid_id(id):
            total += id
print(total)
```
