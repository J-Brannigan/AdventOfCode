```python
import numpy as np
```

```python
with open('puzzle_input.txt') as f:
    rows = [line.rstrip() for line in f.readlines()]
```

```python
rows
```

```python
window_size = 3
padding = int((window_size-1)/2)

rows_array_markers=np.zeros((len(rows),len(rows[0])),dtype='U1')

for row_idx,row in enumerate(rows):
    for roll_idx,roll in enumerate(row):
        rows_array_markers[row_idx,roll_idx]=roll

def generate_arrays(rows_array_markers):
    rows_array_new = np.zeros_like(rows_array_markers)
    rows_array_padded_new = np.zeros((len(rows_array_new)+2*int(np.floor((window_size-1)/2)),len(rows_array_new[0])+2*int(np.floor((window_size-1)/2))))

    for row_idx,row in enumerate(rows_array_markers):
        for roll_idx,roll in enumerate(row):
            if roll == '.':
                roll = 0
            else:
                roll =1
            rows_array_new[row_idx,roll_idx] =roll
            rows_array_padded_new[row_idx+int(np.floor((window_size-1)/2)),roll_idx+int(np.floor((window_size-1)/2))]= roll

    return rows_array_new,rows_array_padded_new

rows_array, rows_array_padded = generate_arrays(rows_array_markers)
```

```python
def remove_rolls(rows_array,rows_array_markers,rows_array_padded):
    markers = rows_array_markers.copy()
    count=0
    for row_idx, row in enumerate(rows_array):
        # print(row)
        for roll_idx, roll in enumerate(row):
            window = rows_array_padded[row_idx:row_idx+2*(padding)+1,roll_idx:roll_idx+2*(padding)+1]
            centre_idx = (int((len(window)-1)/2),int((len(window[0])-1)/2))
            before = window[centre_idx]
            window[centre_idx]=0
            window_sum = np.sum(window)
            window[centre_idx]=before
            if window[centre_idx] == 1 and window_sum < 4:
                count+=1
                markers[row_idx,roll_idx]='x'
            window[centre_idx]=before
    return count, markers
```

```python
rows_array_markers_new = rows_array_markers.copy()
stop = False
total_removed = 0
while not stop:
    # print(rows_array_markers_new)
    for row_idx,row in enumerate(rows_array_markers_new):
        for roll_idx,roll in enumerate(row):
            if roll == 'x':
                rows_array_markers_new[row_idx,roll_idx]='.'
    rows_array_new,rows_array_padded_new = generate_arrays(rows_array_markers_new)
    removed, rows_array_markers_new = remove_rolls(rows_array_new,rows_array_markers_new,rows_array_padded_new)
    if removed == 0:
        stop = True
    total_removed +=removed
print(rows_array_markers_new)
print(total_removed)
```
