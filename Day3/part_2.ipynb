```python
import numpy as np
```

```python
with open('sample_input.txt') as f:
    banks = [line.rstrip() for line in f.readlines()]
```

```python
for bank in banks[:1]:
    argsort = np.argsort([int(i) for i in bank])
    print([int(i) for i in bank])
    print(argsort)
```

```python
def get_largest_joltage(bank):
    argsort = np.argsort([int(i) for i in bank])
    if argsort[-1] == len(bank):
        return f'{bank[argsort[-2]]}{bank[argsort[-3]]}'
    else:
        return f'{bank[argsort[-1]]}{bank[argsort[-2]]}'
    
for bank in banks:
    print(get_largest_joltage(bank))
```

```python
count=0
for bank in banks:
    highest_vals=[]
    last_idx=0
    remaining_str=bank
    size = 12
    for i in range(size):
        characters_remaining=size-i
        highest_val=(max(list(remaining_str[:len(remaining_str)-characters_remaining+1])))
        last_idx=np.argmax(list(remaining_str[:len(remaining_str)-characters_remaining+1]))
        remaining_str=remaining_str[last_idx+1:]
        highest_vals.append(highest_val)
    count+=(int(''.join(highest_vals)))
print(count)
```
