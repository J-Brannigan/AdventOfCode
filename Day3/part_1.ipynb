```python
import numpy as np
```

```python
with open('sample_input.txt') as f:
    banks = [line.rstrip() for line in f.readlines()]
```

```python
banks
```

```python
for bank in banks[:1]:
    argsort = np.argsort([int(i) for i in bank])
    print([int(i) for i in bank])
    print(argsort)
```

```python
def get_largest_joltage(bank):
    argsort = np.argsort([int(i) for i in bank])
    if argsort[-1] == len(bank):
        return f'{bank[argsort[-2]]}{bank[argsort[-3]]}'
    else:
        return f'{bank[argsort[-1]]}{bank[argsort[-2]]}'
    
for bank in banks:
    print(get_largest_joltage(bank))
```

```python
def get_highest_jolts(bank):
    highest_val = max(bank[:-1])
    highest_idx = np.argmax(list(bank[:-1]))
    second_highest_val = max(bank[highest_idx+1:])
    return int(f'{highest_val}{second_highest_val}')
```

```python
counter = 0
for bank in banks:
    counter += get_highest_jolts(bank)
print(counter)
```
