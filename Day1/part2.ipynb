```python
with open("sample_input.txt") as f:
    lines = f.readlines()
    
```

```python
rotations=[]
for line in lines:
    line = line.replace('/n','')
    rotations.append({'direction':line[0],'distance':int(line[1:])})
```

```python
def move_lock(current_location, direction, distance):
    if direction == 'R': sign=1
    elif direction == 'L': sign=-1
    
    new_location = (current_location + sign*distance)%100
    return new_location
```

```python
location = 50
location_history=[location]
for idx,rotation in enumerate(rotations):
    location = move_lock(location,rotation['direction'],rotation['distance'])
    location_history.append(location)
```

```python
location
```

```python
from math import floor

count=0
for i in range(len(location_history)-1):
    print(f'current location: {location_history[i]}')
    print(f'rotation: {rotations[i]}')
    print(f'new location: {location_history[i+1]}')
    full_rotations = floor(rotations[i]['distance']/100)
    if rotations[i]['distance']/100>1:
        print(f"distance: {rotations[i]['distance']}, full rotations: {full_rotations}")
        count += full_rotations
    if location_history[i]+(rotations[i]['distance']%100)>99 and rotations[i]['direction']=='R' and location_history[i]!=0:
        count+=1     
    elif location_history[i]-(rotations[i]['distance']%100)<0 and rotations[i]['direction']=='L' and location_history[i]!=0:
        count+=1
    elif location_history[i+1]==0:
        count+=1
    print(f'count:{count}')
print(count)
```
